```python
url = 'https://www.ycombinator.com/companies/invert/jobs/V9AGYSi-bioprocess-data-scientist'
https://www.ycombinator.com/companies/invert/jobs/V9AGYSi-bioprocess-data-scientist
```

```python
dom = bs4.BeautifulSoup(html, "html.parser")
type(dom.select('body'))
```
