## Modules

```python
import pickle #reads the file
import numpy as np # makes matrixes and vectors speedy
import matplotlib.pyplot as plt #this displays the image
```

## Import Data



* Load pickled data from cats_dogs.pickle



```python
cats_dogs_pickle = '/content/X.pickle' #this line assigns the path to the pickle file to a variable.
dataset = pickle.load(open(cats_dogs_pickle,'rb'))
len(dataset) #this is how many images are in the dataset. it should be equal to 24946
```

## Fit the images and select a random image



* Scale the encoded images so that they fall between 0 and 1 by dividing by 255.

```python
dataset = (dataset)/255
```

## Display a randomly selected Image


```python
random_index = np.random.randint(len(dataset)) #Generates a random number from the datset
plt.imshow(dataset[random_index], 'Greys') ; #Display's the image
plt.title(random_index); #Gives the image a title

# The number in the title coorespods to the index in the dataset of this image.
# In the cell below you can choose this image without the index at the top.
```
