## Modules

```python
import pickle
import pandas as pd
import numpy as np
from PIL import Image
import imageio.v2 as img
import matplotlib.pyplot as plt
import matplotlib.axes
```

## Import Data



* Load pickled data from cats_dogs.pickle



```python
cats_dogs_pickle = '/content/X.pickle'
dataset = pickle.load(open(cats_dogs_pickle,'rb'))
len(dataset) #this is how many images are in the dataset.
```

## Fit the images and select a random image



* Scale the encoded images so that they fall between 0 and 1 by dividing by 255.

```python
dataset = (dataset)/255
```

## Display a randomly selected Image


```python
random_index = np.random.randint(len(dataset))
plt.imshow(dataset[random_index], 'Greys') ;
plt.title(random_index);

# The number in the title coorespods to the index in the dataset of this image.
# In the cell below you can set choose recreate this image with out the index at the top.
```
